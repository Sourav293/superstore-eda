## Insight 2 — Category & Sub-Category Profitability

```python
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x='Category', y='Profit', data=profit_by_category, color='steelblue')
plt.title('Profit by Category')
plt.show()
```

- Technology is the most profitable category
- Office Supplies follows with steady profit
- Furniture is the least profitable category

**Sub-Category level:**

```python
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values('Profit', ascending=False).head(10)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Sub-Category', y='Profit', data=profit_by_subcategory, color='steelblue')
plt.title('Profit by Sub-Category')
plt.xticks(rotation=45, ha='right')
plt.show()
```

- Copiers, Phones, Accessories are top performers
- Tables are making a significant LOSS
- Bookcases and Supplies are also loss-making
- Recommendation: business should review pricing
  strategy for Tables and Bookcases — consider
  discontinuing or repricing these products

## Insight 3 — Discount vs Profit Relationship

```python
plt.figure(figsize=(8,4))
plt.scatter(df['Discount'], df['Profit'], alpha=0.5, color='steelblue')
plt.title('Profit vs Discount')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

```python
profit_by_discount = df.groupby('Discount')['Profit'].mean().reset_index().sort_values('Profit', ascending=False)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Discount', y='Profit', data=profit_by_discount, color='steelblue')
plt.title('Profit at different Discount level')
plt.show()
```


- Clear negative relationship between discount and profit
- At 0% discount — healthy profit margins
- After 20% discount — profit turns negative (loss)
- Heavy discounting (40-80%) causes significant losses
- Recommendation: cap discounts at 20% maximum
  to protect profit margins

## Insight 4 — Customer Segment Analysis

```python
sales_per_segment = df.groupby('Segment')['Sales'].sum().reset_index()
profit_per_segment = df.groupby('Segment')['Profit'].sum().reset_index()
discount_per_segment = df.groupby('Segment')['Discount'].mean().reset_index()
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,5))

sns.barplot(x='Segment', y='Sales', data=sales_per_segment, color='steelblue', ax=axes[0])
axes[0].set_title('Sales Per Segment')

sns.barplot(x='Segment', y='Profit', data=profit_per_segment, color='coral', ax=axes[1])
axes[1].set_title('Profit Per Segment')

sns.barplot(x='Segment', y='Discount', data=discount_per_segment, color='orange', ax=axes[2])
axes[2].set_title('Discount Per Segment')

plt.tight_layout()
plt.show()
```

- Consumer segment generates highest sales and profit
- Corporate segment has similar discount level as Consumer
  but lower sales and profit — smaller order volumes
- Home Office has similar discount rate but lowest
  sales and profit overall
- Discount is not the differentiating factor between segments
- Recommendation: focus marketing efforts on Consumer
  segment — highest return on investment

## Insight 5 — Monthly Sales & Profit Trend

```python
sale_by_month = df.groupby('Order Month')['Sales'].sum().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(x='Order Month', y='Sales', marker='o', data=sale_by_month, color='steelblue')
plt.title('Sales by Month')
plt.show()
```

```python
profit_by_month = df.groupby('Order Month')['Profit'].sum().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(x='Order Month', y='Profit', marker='o', data=profit_by_month, color='steelblue')
plt.title('Profit by Month')
plt.show()
```

- March (Month 3) shows significant profit spike
- April (Month 4) sees sharp decline in both sales and profit
- Months 5-8 show steady but modest performance
- September onwards shows strong recovery
- December (Month 12) peaks in both sales and profit
  — likely holiday season effect
- Recommendation: increase inventory and marketing
  budget in Q4 (Oct-Dec) to capitalize on peak season
  and investigate April decline

## Insight 7 — State Level Profitability


```python
top5_state_by_profit = df.groupby('State')['Profit'].sum().reset_index().sort_values('Profit', ascending=False).head(5)
least5_state_by_profit = df.groupby('State')['Profit'].sum().reset_index().sort_values('Profit', ascending=True).head(5)

fig, axes = plt.subplots(1,2, figsize=(18,6))

sns.barplot(x='State', y='Profit', data=top5_state_by_profit, color='coral', ax=axes[0])
axes[0].set_title('Top 5 States by Profit')

sns.barplot(x='State', y='Profit', data=least5_state_by_profit, color='green', ax=axes[1])
axes[1].set_title('Bottom 5 states by Profit')

plt.tight_layout()
plt.show()
```

- California and New York are top performing states
  — likely due to large population and high order volume
- Texas and Ohio are making significant losses
  — combined with Insight 3 (discounting),
  heavy discounts in these states may be the cause
 - Recommendation: immediately cap discounts at 20%
  in Texas and Ohio — this single action could
  turn these states from loss-making to profitable

## Insight 8 — Shipping Mode Analysis

```python
sns.countplot(x='Ship Mode', data=df)
```

```python
profit_per_shippingmode = df.groupby('Ship Mode')['Profit'].sum().reset_index()
plt.figure(figsize=(10,5))
sns.barplot(x='Ship Mode', y='Profit', data=profit_per_shippingmode, color='steelblue')
plt.title('Profit by Different Mode')
plt.show()
```

```python
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
avg_shipping_days_by_shipsMode = df.groupby('Ship Mode')['Shipping Days'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.barplot(x='Ship Mode', y='Shipping Days', data=avg_shipping_days_by_shipsMode, color='steelblue')
plt.title('Average Shipping Days By Diff Ship Mode')
plt.show()
```

- Standard Class is most used (approx 60% of orders)
  despite taking 5 days average delivery time
- Same Day delivery takes 0 days as expected
- First Class (2 days) and Second Class (3 days)
  are middle ground options
- Standard Class generates highest profit due to
  high order volume despite lower margins per order
- Recommendation: promote First Class shipping
  as an upsell option — faster delivery at
  slightly higher cost could improve margins

## Insight 9 — Correlation Analysis

```python
plt.figure(figsize=(8, 6))
correlation = df[['Sales', 'Profit', 'Discount', 'Quantity', 'Shipping Days']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

- Discount vs Profit = -0.22 (negative correlation)
  confirms Insight 3 — higher discount = lower profit
  
- Sales vs Profit = 0.48 (moderate positive)
  higher sales generally leads to higher profit
  but not always — discounting breaks this relationship
  
- Quantity vs Profit = 0.07 (very weak positive)
  selling more units does not guarantee more profit
  product margins matter more than quantity
